```python
def normalize_probs(prob_1, prob_2):
    total = prob_1 + prob_2
    return prob_1/total, prob_2/total
```

```python
# Funció simple:
def point_sim(server_pct, receiver_pct):
    server_pct, receiver_pct = normalize_probs(server_pct, receiver_pct)
    x = random.random()
    if x < server_pct:
        return 1
    else:
        return 0
```

```python
def tiebreak_sim(player_a, player_b):
    points_a = 0
    points_b = 0
    switch = True
    while points_a < 7 and points_b < 7:
        if (points_a + points_b) % 2 == 1:
            switch = not(switch)
        if switch == 0:
            a_pct = player_a['player_break_points_saved', 'sum'].values[0]/player_a['player_break_points_serve_total', 'sum'].values[0]
            b_pct = player_b['player_break_points_converted', 'sum'].values[0]/player_b['player_break_points_return_total', 'sum'].values[0]
            point = point_sim(a_pct,b_pct)
            if point == 1:
                if points_a == 5 and points_b == 6:
                    points_b -= 1
                else: 
                    points_a += 1
            elif point == 0:
                if points_a == 6 and points_b == 5:
                    points_a -= 1
                else: 
                    points_b += 1
#             print('Switch: ', switch, 'Points_A: ', points_a, 'Points B: ', points_b, 'Pooint: ', point)
        elif switch == 1:
            a_pct = player_a['player_break_points_converted', 'sum'].values[0]/player_a['player_break_points_return_total', 'sum'].values[0]
            b_pct = player_b['player_break_points_saved', 'sum'].values[0]/player_b['player_break_points_serve_total', 'sum'].values[0]
            point = point_sim(b_pct, a_pct)
            if point == 1:
                if points_b == 5 and points_a == 6:
                    points_a -= 1
                else:
                    points_b += 1
            elif point == 0:
                if points_b == 6 and points_a == 5:
                    points_b -= 1
                else:
                    points_a += 1
#             print('Switch: ', switch, 'Points_A: ', points_a, 'Points B: ', points_b, 'Pooint: ', point)
            
    if points_a > points_b:
        return 1
    else:
        return 0
    
```

```python
# https://towardsdatascience.com/building-a-tennis-match-simulator-in-python-3add9af6bebe
# S'utilitza la lògica del sistema de punts d'aquesta pàgina.

def game_sim(server,receiver): 
    server_points = 0
    receiver_points = 0
    server_prob = ((server['player_first_serves_in', 'sum'].values[0] / server['player_first_serves_total', 'sum'].values[0])
    *(server['player_first_serve_points_won', 'sum'].values[0]/ server['player_first_serve_points_total', 'sum'].values[0])
    +(1 - server['player_first_serves_in', 'sum'].values[0] / server['player_first_serves_total', 'sum'].values[0])
    *(1 - server['player_double_faults', 'sum'].values[0]/ server['player_second_serve_points_total','sum'].values[0])
    *(server['player_second_serve_points_won','sum'].values[0]/server['player_second_serve_points_total','sum'].values[0]))
         
    receiver_prob = ((server['player_first_serves_in', 'sum'].values[0] / server['player_first_serves_total', 'sum'].values[0])
    *(receiver['player_first_serve_return_won','sum'].values[0]/ receiver['player_first_serve_return_total','sum'].values[0])
    +(server['player_double_faults', 'sum'].values[0]/ server['player_second_serve_points_total','sum'].values[0])
    +(1 - (server['player_first_serves_in', 'sum'].values[0] / server['player_first_serves_total', 'sum'].values[0]))
    *(receiver['player_second_serve_return_won', 'sum'].values[0]/ receiver['player_second_serve_return_total','sum'].values[0]))
#     server_prob = server
#     receiver_prob = receiver
    
    while server_points < 4 and receiver_points < 4: 
        point = point_sim(server_prob, receiver_prob)
        if point == 1:
            if server_points == 2 and receiver_points == 3:
                receiver_points -= 1
            else: 
                server_points += 1
        else:
            if server_points == 3 and receiver_points == 2:
                server_points -= 1
            else:
                receiver_points += 1
#         print('Server points:', server_points, ' Receiver points:', receiver_points, ' Point:', point)

#         BREAKPOINT SITUATION

        if receiver_points == 3:
            server_break = server['player_break_points_saved', 'sum'].values[0]/server['player_break_points_serve_total', 'sum'].values[0]
            receiver_break = receiver['player_break_points_converted', 'sum'].values[0]/receiver['player_break_points_return_total', 'sum'].values[0]
            point = point_sim(server_break,receiver_break)
#             point = point_sim(server, receiver)
            if point == 1:
                if server_points == 2:
                    receiver_points -= 1
                else: 
                    server_points += 1
            else:
                receiver_points += 1
                
#             print('Server points:', server_points, ' Receiver points:', receiver_points, ' Point:', point)


    if server_points == 4: 
        return 1
    elif receiver_points == 4:
        return 0
```

```python
def set_sim(player_a, player_b):
    games_a = 0
    games_b = 0
    games = 0
    while games_a < 6 and games_b < 6:
        if games%2 == 0:
#             print('Server player A - Receiver player B')
            game = game_sim(player_a, player_b)
            if game == 1:
                games_a += 1
            else:
                games_b +=1
        else:
#             print('Server player B - Receiver player A')
            game = game_sim(player_b, player_a)
            if game == 1:
                games_b += 1
            else:
                games_a +=1
        games += 1
#         print('Game A:', games_a, '------ Games B:', games_b, ' Game:', game)
    if abs(games_a - games_b) < 2:
        if games%2 == 0:
#             print('Server player A - Receiver player B')
            game = game_sim(player_a, player_b)
            if game == 1:
                games_a += 1
            else:
                games_b +=1
        else:
#             print('Server player B - Receiver player A')
            game = game_sim(player_b, player_a)
            if game == 1:
                games_b += 1
            else:
                games_a +=1
#         print('Game A:', games_a, '------ Games B:', games_b)
        if games_a == games_b:
#             print('tiebreak')
            tiebreak = tiebreak_sim(player_a, player_b)
            if tiebreak == 1:
                return 1
            else:
                return 0
        elif games_a > games_b:
            return 1
        elif games_b > games_a:
            return 0
    else:
        if games_a == 6:
            return 1
        elif games_b == 6:
            return 0
```

```python
def match_sim(player_a, player_b, num_set):
    sets_a = 0
    sets_b = 0 
    sets_played = 0
    while sets_played < num_set and (sets_a < ((num_set//2) +1) or sets_b < ((num_set//2) +1)):
        set_result = set_sim(player_a, player_b)
        if set_result == 1:
            sets_a += 1
        else:
            sets_b += 1
        sets_played += 1
#         print('Sets A: ', sets_a, ' Sets B: ', sets_b, 'Sets Played: ', sets_played)
#         print('---------------------------------------------------------------------')
    if sets_a > sets_b:
        return 1
    elif sets_b > sets_a:
        return 0
```
